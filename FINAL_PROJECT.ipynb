### Transportation

#**Dataset Information:**

**Source:** Government portal / Data.gov /  Data Portal link-https://catalog.data.gov/dataset/state-level-transportation-public-finance-statistics-tpfs

**Year / Timeline:** Data updated on Feb 6,2026

**Dataset Column/features Description:**

* **Cash_flow** -Indicates whether the record represents Revenue (money received) or Expenditure (money spent).

* **Own_supporting** - Transportation activities are financially self-supporting or supported by other government funds.

* **User_other** - Identifies whether revenue comes from users (fees, tolls, fares) or other sources (taxes, grants, transfers).

* **Trust_fund** - Whether the transaction is associated with a Trust Fund or Non-Trust Fund.

* **Govt_level** - Level of government responsible for the transaction (State, Local, or State and Local combined).

* **Description**- Detailed explanation of the revenue or expenditure category.


* **State** - Name of the U.S. state.

* **Year** - Reporting year of the record.

* **Value** - Original amount reported for the category
* **Mode** - Transportation mode associated with the record (e.g., highways, transit, aviation, rail).


* **Deflator** - Inflation adjustment factor

* **Chained** - Inflation-adjusted value using the deflator.

* **Population** - Population of the state in the given year.

* **Value per population** - Per-capita value calculated as: Value ÷ Population.
* **Chained per population** - Inflation-adjusted per-capita value calculated as: Chained Value ÷ Population.

# **DROP**

```python
df.drop(['exp_type','state_code','gov_level','mode_sort_order','state_code_sort_order','gov_level_sort_order','user_other','estimate_actual','user_other_grp','chained_year'],axis=1,inplace=True)
```

```python
df.drop(df[df['year']==2023].index,inplace=True)
```

```python
df.drop(df[df['mode']=='Total'].index,inplace=True)
```

## **FILLING NULL VALUES**

```python
(df['deflator'].fillna(df['deflator'].mean(),inplace=True))
(df['chained'].fillna(df['chained'].mean(),inplace=True))
(df['chained_per_pop'].fillna(df['chained_per_pop'].mean(),inplace=True))
```

## **RENAMING**

```python
df.rename(columns={'trust_fund':'fund_type','own_supporting':'supporting_type'},inplace=True)
```

## **REPLACE**

```python
df['supporting_type']=df['supporting_type'].replace({'Total':'Other Supporting'})
```

```python
df['fund_type']=df['fund_type'].replace({'Total':'trust fund'})
```

```python
df['description']=df['description'].replace({'-':'other'})
```

```python
numerical_col=df.select_dtypes(include=['int64','float64']).columns

for col in numerical_col:
  mean_val=df.loc[df[col] !=0,col].mean()
  df[col]=df[col].replace(0,mean_val)
```

```python
df['value_crore']=(df['value']/10000000).round(2)
df['population_crore']=(df['population']/10000000).round(5)
```

## **FIND THE OUTLIERS**

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_col=df.select_dtypes(include=['int64','float64']).columns

for col in numerical_col:
  plt.figure(figsize=(7,4))
  sns.boxplot(x=df[col])
  plt.title(f'boxplot of{col}')
  plt.show()
```

## **NO OF OUTLIERS (BEFORE TREATMENT)**

```python
for col in numerical_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f'{col}: {len(outliers)} outliers')
```

```python
print(df.shape)
```

## **Removing Outliers**

```python
for col in numerical_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df=df[(df[col]>=lower)&(df[col]<=upper)]
```

```python
print(df.shape)
```

## **NO OF OUTLIERS (AFTER TREATMENT)**

```python
for col in numerical_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

**Wisconsin and North Carolina have the lowest values.**

**Most states fall within the range of 45–65.**
